### Video 21 Preparing the dataset

```python
import pandas as pd
import bar_chart_race as bcr
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("time_series_covid19_confirmed_global.csv")
df.head()
```

```python
df.drop(["Lat", "Long"], axis = "columns", inplace=True)
df.head()
```

```python
# Remove repeted Country/Region by summing up it

df = df.groupby("Country/Region").sum()
df.head(10)
```

```python
# Changing the rows to columns

df = df.T
df.head()
```

```python
# Set the index as a datetime index
df.index = pd.to_datetime(df.index)
df.index
```

```python
# Create the barchart
bcr.bar_chart_race(
    df = df, 
    filename = "covid19_final.mp4",
    orientation = 'h',
    sort = 'desc',
    n_bars = 10,
    fixed_order = False,
    fixed_max = False,
    steps_per_period = 10,
    period_length = 250,
    interpolate_period = False,
    label_bars = True,
    bar_size = 0.95,
    period_label = True, 
    period_fmt = "%B %d, %Y",
    period_summary_func = lambda v, r: {'x': .99, 'y': .3,
                                      's': f'Total cases: {v.nlargest(6).sum():,.0f}',
                                      'ha': 'right', 'size': 9, 'family': 'Courier New'},
    perpendicular_bar_func = None,
    figsize = (6, 3.5),
    cmap = None,
    title = "Top 10 countries - Covid19 Cases",
    title_size = None,
    bar_label_size = 7,
    tick_label_size = 7,
    shared_fontdict = None,
    scale = 'linear',
    writer = None,
    fig = None,
    dpi = 300,
    bar_kwargs = None,
    filter_column_colors = False
)
```
